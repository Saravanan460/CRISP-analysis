```python
import pandas as pd

# Load the data
df = pd.read_csv(r'data sets\2019Floor2.csv')

# Show basic info
print(df.describe())


```

```python
# Check for missing values
print(df.isnull().sum())

# Check time format
print(df.columns[0])  # usually the timestamp column
print(df[df.columns[0]].head())

# Convert timestamp if needed
df['Timestamp'] = pd.to_datetime(df[df.columns[0]])
df.set_index('Timestamp', inplace=True)
```

```python
import matplotlib.pyplot as plt

df['z2_AC1(kW)'].plot(figsize=(15, 5), title='Zone 2 AC Unit 1 Usage (kW)')
plt.ylabel('Power (kW)')
plt.show()
df['z1_S1(degC)'].plot(figsize=(15, 5), title='Zone 1 temp')
plt.ylabel('temp (degC)')
plt.show()
df['z2_S1(degC)'].plot(figsize=(15, 5), title='Zone 2 temp')
plt.ylabel('temp (degC)')
plt.show()
df['z3_S1(degC)'].plot(figsize=(15, 5), title='Zone 3 temp')
plt.ylabel('temp (degC)')
plt.show()
df['z4_S1(degC)'].plot(figsize=(15, 5), title='Zone 4 temp')
plt.ylabel('temp (degC)')
plt.show()
```
